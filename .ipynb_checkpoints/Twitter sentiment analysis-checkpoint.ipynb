# Create a pipeline to remove stop-words, punctuation, and perform tokenization

```python
#defining a pipeline to clean up all the messages
#pipeline performs 1.) remove punctuations, 2.) remove stopwords
def message_cleaning(message):
    punc_removed = [char for char in message if char not in string.punctuation]
    punc_join = ''.join(punc_removed)
    punc_join_clean = [word for word in punc_join.split() if word.lower() not in stopwords.words('english')]
    return punc_join_clean
```

```python
tweets_df_clean = tweets_df['tweet'].apply(message_cleaning)
```

```python
tweets_df_clean
```

```python
#Define the cleaning pipeline we defined earlier
vectorizer = CountVectorizer(analyzer = message_cleaning)
```

```python
tweets_countvectorizer = CountVectorizer(analyzer=message_cleaning, dtype='uint8').fit_transform(tweets_df['tweet']).toarray()
```

```python
tweets_countvectorizer.shape
```

```python
x = tweets_countvectorizer
x.shape
```

```python
y = tweets_df['label']
y.shape
```

# Training a Naive Bayes Classifier

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test = NB_classifier.predict(x_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```
